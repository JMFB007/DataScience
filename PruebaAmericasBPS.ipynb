# 4. Usando la librería Pandas, leer el archivo CSV "deudores.csv" y almacenarlo en una variable manteniendo la información lo más fiel posible a la fuente.

- Imprimir el DataFrame inicial en consola.

```python
deudores = pd.read_csv('deudores.csv', sep=';',encoding="utf-8-sig")
deudores
```

**Realizar las siguientes transformaciones:**

- Convertir los datos de la columna "Préstamo en COP" de flotante a entero.

```python
deudores["Préstamo en COP"] = deudores["Préstamo en COP"].astype(int)
deudores
```

- Agregar una columna "Contribución mínima mensual", la cual es la división entre "Préstamo en COP" y "Plazo de pago en meses". Los datos de la columna deben ser tipo entero.

```python
deudores["Contribución mínima mensual"] = (deudores["Préstamo en COP"]/deudores["Plazo de pago en meses"]).astype(int)
deudores
```

- Reemplazar las "M" por "Masculino" y las "F" por "Femenino" en la columna "Sexo".

```python
deudores['Sexo'] = deudores['Sexo'].replace(['M'], 'Masculino')
deudores['Sexo'] = deudores['Sexo'].replace(['F'], 'Femenino')
deudores["Sexo"] = deudores["Sexo"].apply(lambda x: "Masculino" if x == "M" else "Femenino")
```

- Imprimir el DataFrame resultante.

```python
deudores
```

- Filtrar las mujeres con 25 años o más y almacenarlas en una variable nueva "objetivo".

```python
objetivo = deudores[deudores['Edad'] >= 25][deudores['Sexo'] == "Femenino"]
objetivo
objetivo = deudores.loc[(deudores["Edad"] >= 25) & (deudores["Sexo"].str.contains("Femenino"))]
```

- Cambiar el nombre de "Valeria Paredes" a "Luisa Restrepo" usando loc y condicionales. El cambio se debe guardar en la misma variable "objetivo".

```python
#mejor metodo
#   objetivo['Nombre'] = objetivo['Nombre'].replace(['Valeria Paredes'], 'Luisa Restrepo')
for i in objetivo.index.values.tolist():
  if objetivo.loc[i,'Nombre'] == "Valeria Paredes":
    objetivo.loc[i,'Nombre'] = "Luisa Restrepo"
objetivo

objetivo.loc[objetivo["Nombre"].str.contains("Valeria Paredes"),"Nombre"] = "Luisa Restrepo"
```

- Ordenar el DataFrame por "Contribución mínima mensual" en orden descendente.

```python
objetivo = objetivo.sort_values(by=["Contribución mínima mensual"], ascending=False)
objetivo
```

- Agrupar el DataFrame con la columna "Sexo". Se debe imprimir el promedio de "Edad", la suma de "Préstamo en COP", la suma de "Plazo de pago en meses" y la suma de "Contribución mínima mensual".

```python
d = {"Edad":'Edad Promedio',"Préstamo en COP":'Suma Préstamo en COP',"Plazo de pago en meses":'Suma Plazo de pago en meses',"Contribución mínima mensual":'Suma Contribución mínima mensual'}
group = objetivo.groupby(["Sexo"]).agg({"Edad":'mean', "Préstamo en COP":'sum',"Plazo de pago en meses":'sum',"Contribución mínima mensual":'sum'}).rename(columns=d)
group
```
