```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv('SalarioMinimo.csv', index_col=None).iloc[:-1,:-1]
df = df.iloc[::-1]
df.head()
# from matplotlib import ticker

X = np.array(df["Año"]).reshape(-1, 1)
Y = np.array(df["SalarioMinimoMensual"]).reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)

lm = LinearRegression()
lm.fit(x_train,y_train)
y_pred = lm.predict(x_test)
poly = PolynomialFeatures(degree=2, include_bias=False)

poly_features = poly.fit_transform(X)

poly_reg_model = LinearRegression()
poly_reg_model.fit(poly_features, Y)

y_predicted = poly_reg_model.predict(poly_features)
fig, ax = plt.subplots()
ax.ticklabel_format(style='plain')
plt.scatter(np.array(df["Año"]), np.array(df["SalarioMinimoMensual"]))
plt.xticks(rotation=90, ha="right")
plt.grid()
plt.plot(np.array(df["Año"]), y_predicted, color='purple')
```

```python
predList = [[2026]]
poly_features2 = poly.fit_transform(np.array(predList).reshape(-1, 1))
y_predicted = poly_reg_model.predict(poly_features2)
print(y_predicted)
plt.xticks(rotation=90, ha="right")
plt.scatter(np.array(predList), y_predicted, color='purple')
```
