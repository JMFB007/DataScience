Ejercicio de pandas y numpy para procesar y graficar datos

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy

df = pd.read_csv(r"C:\Users\FeJo433\Desktop\StoresData.csv")
df.head()
```

```python
df["Average_Sales_Per_Customer"] = df["Store_Sales"]/df["Daily_Customer_Count"]
df.sort_values(by=['Daily_Customer_Count'], ascending=False)
dfg = df.groupby(["Store ID ", "Store_Area"]).sum().sort_values(by=['Daily_Customer_Count'], ascending=False)
display(dfg)
```

```python
top10 = df.sort_values(by='Daily_Customer_Count', ascending=True).head(10)
top10['Store ID '] = top10['Store ID '].astype(str)
display(top10)
plt.figure(figsize=(10,8))
plt.bar(
  x=top10['Store ID '],
  height=top10['Daily_Customer_Count'],
  label="Ventas mensuales",
  color="skyblue")

plt.title("Average Daily Customer Count per Store ID")
plt.xlabel("Store ID")
plt.ylabel("Daily Customer Count")
plt.legend()
plt.show()
```
