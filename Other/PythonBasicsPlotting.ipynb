```python
import matplotlib.pyplot as plt
import pandas as pd
import csv

x = pd.read_csv("data.csv")
def barchar(labels, values):
  plt.pie(x.loc[:,"World Population Percentage"], labels = x.loc[:,"Country"])
  plt.axis("equal")
  plt.show()

if __name__ == "__main__":
  labels = ["a", "b", "c"]
  values = [1, 2, 3]
  barchar(labels, values)
```
